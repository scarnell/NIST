```python
import pandas as pd
from scipy.spatial import distance as dist
from datetime import timedelta
import numpy as np
from __future__ import division
```

```python
def getTimeStamp(timeStamp):
    return pd.to_datetime(timeStamp)

def createDF(folderName):
    flow_df = pd.read_csv(folderName + 'flow.tsv', delimiter='\t', header=None)
    speed_df = pd.read_csv(folderName + 'speed.tsv', delimiter='\t', header=None)
    occupancy_df = pd.read_csv(folderName + 'occupancy.tsv', delimiter='\t', header=None)
    
    df = pd.DataFrame({'A' : []})
    del df['A']
    for col in flow_df:
        df[col] = pd.Series(data = list(zip(flow_df[col],speed_df[col], occupancy_df[col])))
    
    timeStamp_df = pd.read_csv(folderName + 'timestamp.tsv', delimiter='\t', header=None)
    df['timeStamp'] = getTimeStamp(timeStamp_df[0])
    return df
```

```python
def containsNeg(vec):
    #flow
    if(vec[0] < 0 or vec[0] > 250):
        return 1
    #speed
    elif(vec[1] < 0):
        return 1
    #occupancy
    elif(vec[2] < 0):
        return 1
    else:
        return 0
```

```python
def getVecInTimePeriod(timeInterval, currTime, df):
    #print "Curr Time: ",currTime
    timeInterval = timedelta(minutes = timeInterval)
    startTime = currTime - timeInterval
    #print "Start Time: ",startTime
    endTime = currTime + timeInterval
    #print "End Time: ",endTime
    df = df[df['timeStamp'] >= startTime]
    return df[df['timeStamp'] <= endTime]
```

```python
def combineScore(lane_p,zone_p):
    return (LANE_WEIGHT * lane_p) + (ZONE_WEIGHT * zone_p)
```

```python
#same lane calculations
def compareLaneData(currCol,currVec,df, distThreshold):
    row_index = df.index[0]
    num_rows = df.shape[0]
    max_index = row_index + num_rows
    numVecGreater = 0
    totalVec = 0
    while row_index < max_index:
        #print(containsNeg(df[currCol][row_index]))
        if(containsNeg(df[currCol][row_index]) != 1):
            edist = dist.euclidean(df[col][row_index], currVec)
            if(edist > distThreshold):
                numVecGreater += 1
        totalVec += 1
        row_index += 1
        #print("current vec " + str(currVec) + " compared to " + str(df[currCol][row_index]))
    if(totalVec != 0):
        return numVecGreater/totalVec
    else:
        return 0

#same zone calculations
def compareZoneData(currCol,currVec, df, distThreshold):
    #print("CurrVec: " + str(currVec))
    #print("Current col: "+ str(currCol))
    num_rows=df.shape[0]
    total_cols=df.shape[1]
    check_cols=np.arange(total_cols-1)
    check_cols=np.delete(check_cols,currCol)
    row_index = df.index[0]
    max_index = row_index + num_rows
    col_index = 0
    numVecGreater = 0
    totalVec=0
    for col in check_cols:
        while row_index < max_index:
            if(containsNeg(df[currCol][row_index]) == 0):
                edist = dist.euclidean(df[col][row_index], currVec)
                #print("Row ",row_index," Col ",col," dist: ",edist)
                if(edist > distThreshold):
                    numVecGreater += 1
            totalVec += 1
            row_index += 1
        row_index = df.index[0]
    #print("Total Vectors: " + str(totalVec))
    #print("Num Vectors (>): " + str(numVecGreater))
    if(totalVec != 0):
        return numVecGreater/totalVec
    else:
        return 0
```

# RUN THIS TO GET FINAL PROBS FOR A SINGLE ZONE (FOLDER). MAKE LOOP TO DO ALL. NO IDEA HOW LONG THIS WILL RUN FOR. NOTE: OPTIMIZE THRESHOLDS FIRST BEFORE RUNNING ALL? NEED TO WRITE A TEXT FILE OUTPUT FUNCTION

```python
#1160
zones = [3451, 3232, 1160, 3445, 3532]
```

```python
TIME_INT = 5 #mins
THRESH_DET=6.5
THRESH_ZONE=15
LANE_WEIGHT = 0.85
ZONE_WEIGHT = 0.15
neg_counter=0
totalVec = 0
greaterThan = [0,0,0,0]

for zid in zones:
    zone_id = zid
    print("RUNNING ZONE " + str(zone_id))
    zone = createDF('cleaning1/' + str(zone_id) + "/")
    
    NUM_ROWS = zone.shape[0]
    NUM_COLS = zone.shape[1] - 1

    output = open(str(zone_id) + ".txt", 'w')
    output2 = open(str(zone_id) + "_format.txt", 'w')

    print("Num rows: " + str(NUM_ROWS))
    #print "Num cols: ",NUM_COLS
    for row in np.arange(NUM_ROWS):
        currTime=zone['timeStamp'][row]
        snippet = getVecInTimePeriod(TIME_INT, currTime, zone) #kill 2 birds at once, do all vectors on row before grabbing new time group
        rowResult = ""
        for col in np.arange(NUM_COLS):
            currCol=col
            currRow=row
            currVec=zone[currCol][currRow]
            if(containsNeg(currVec) == 1):
                comb_p = 1.0
            else:
                lane_p=compareLaneData(currCol,currVec, snippet, THRESH_DET)
                zone_p=compareZoneData(currCol,currVec, snippet, THRESH_ZONE)
                comb_p = combineScore(lane_p,zone_p)
                comb_p = round(comb_p, 8)
            #print "Comb (p): ",comb_p
            result = str(currVec[0]) + "\t" + str(currVec[1]) + "\t" + str(currVec[2]) + "\t" + str(comb_p)
            comb = result + "\n"

            if(comb_p > 0.9):
                greaterThan[0] += 1
            if(comb_p > 0.8):
                greaterThan[1] += 1
            if(comb_p > 0.7):
                greaterThan[2] += 1
            if(comb_p > 0.6):
                greaterThan[3] += 1

            output.write(comb)
            rowResult += "("+ str(round(currVec[0], 0))+ ", " + str(round(currVec[1], 0))+ ", " + str(round(currVec[2], 0)) +")\t" + str(comb_p) + "\t\t\t"
            totalVec += 1
        rowResult += "\n"
        output2.write(rowResult)
        if(row %100 == 0):
            print("ran row " + str(row))
            
    print('total number of vectors ' + str(totalVec))
    currThreshold = 0.9
    output3 = open(str(zone_id) + "_stats.txt", 'w')
    for measure in greaterThan:
        percent = round((measure / totalVec)*100, 3)
        output3.write("vec greater than " + str(currThreshold) + ": " + str(percent) + "%")
        currThreshold = round(currThreshold - 0.1,1)
    output.flush()
    output2.flush()
    output3.flush()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#VIZ DATA - To get a feel for the distribution of spd,occup, and flow values for x minus
ROW=100 #specific data point's row
COL=0 #specific data point's col
NUM_BINS=5 #histogram
NUM_PLOTS=4 #dont change
currVec=zone[COL][ROW] #curr vector
snippet = getVecInTimePeriod(TIME_INT, zone['timeStamp'][ROW], zone) #grabs time segment

#within detector metrics
def plotLaneHist(currCol,currVec, df):
    print("CURR VEC: " + str(currVec))
    vals=np.arange(3) #iterator for columns
    row_index = 0
    num_rows = df.shape[0]
    max_index = row_index + num_rows
    col_index = 0
    plot=1; #for plt.subplot
    for measure in vals: #flow,spd,occup
        while row_index < max_index:
            loc=df[currCol][row_index][measure] #data value (eg. '67' mph)
            plt.subplot(1,NUM_PLOTS,plot) #plot hist
            fig=plt.hist(loc,bins=NUM_BINS)
            row_index += 1
            plt.ylim(0,5)
            if (plot==1):
                #flow
                plt.xlim(-50,260)
                plt.title('Flow')
                plt.subplot(1,NUM_PLOTS,4)
                all_flow=[x[0] for x in df[currCol][:]]
                plt.plot(np.arange(num_rows),all_flow,color='b') #all plots (4th fig)
                plt.ylim(-50,260)
                plt.title("ALL METRICS")
            if (plot==2):
                #spd
                plt.xlim(-50,150)
                plt.title('Spd')
                plt.subplot(1,NUM_PLOTS,4)
                all_flow=[x[1] for x in df[currCol][:]]
                plt.plot(np.arange(num_rows),all_flow,color='r')
                plt.ylim(-50,260)
                plt.title("ALL METRICS")
            if (plot==3):
                #occup
                plt.xlim(0,100)
                plt.title('Occup')
                plt.subplot(1,NUM_PLOTS,4)
                all_flow=[x[2] for x in df[currCol][:]]
                plt.plot(np.arange(num_rows),all_flow,color='k')
                plt.ylim(-50,260)
                plt.title("ALL METRICS")
        row_index = df.index[0]
        plot += 1

#within zone metrics
def plotZoneHist(currCol,currVec, df):
    print("CURR VEC: " + str(currVec))
    vals=np.arange(3)
    row_index = 0
    num_rows = df.shape[0]
    max_index = row_index + num_rows
    total_cols=df.shape[1]
    check_cols=np.arange(total_cols-1)
    check_cols=np.delete(check_cols,currCol)
    col_index = 0
    plot=1;
    for measure in vals:
        for col in check_cols:
            while row_index < max_index:
                loc=df[col][row_index][measure]
                plt.subplot(1,NUM_PLOTS,plot)
                fig=plt.hist(loc,bins=NUM_BINS)
                row_index += 1
                plt.ylim(0,5)
                if (plot==1):
                    #flow
                    plt.xlim(-50,260)
                    plt.title('Flow')
                if (plot==2):
                    #spd
                    plt.xlim(-50,150)
                    plt.title('Spd')
                if (plot==3):
                    #occup
                    plt.xlim(0,100)
                    plt.title('Occup')
            row_index = df.index[0]
        plot += 1
        row_index = df.index[0]
```
